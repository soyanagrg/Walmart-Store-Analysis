# Walmart Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Store Status Analysis
Question: How many Walmart stores are currently Open, Closed, or in another status?
- Visualization: Bar chart of store status

```python
df_store_status = pd.read_csv("data/Store_Status_Analysis.csv")

bars = plt.bar(df_store_status['Operation_Status'], df_store_status['Store_Count'], color=['green', 'red'])

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, yval, ha='center', va='bottom')

plt.title("Walmart Store Status Analysis")
plt.xlabel("Store Status")
plt.ylabel("Number of Stores")
plt.ylim(0, 6000)
plt.show()
```

Out of 5209 stores, 5208 are operational and only 1 is closed, suggesting this dataset may reflect a specific snapshot in time or only active store records

## 2. Store Dsitribution by State
Question: How are Walmart stores distributed across the United States by state?
- Visualization: Horizontal bar chart of stores per state

```python
df_top_states = pd.read_csv("data/Store_Distribution_By_State_Top10.csv")

top_states = plt.barh(df_top_states['State'], df_top_states['State_Count'], color='steelblue')

for top_state in top_states:
    xval = top_state.get_width()
    plt.text(xval, top_state.get_y() + top_state.get_height()/2, xval, ha='left', va='center')

plt.title("Top 10 Store Distribution by State")
plt.ylabel("State")
plt.xlabel("Number of Stores")
plt.show()
```

The top 10 states with the most Walmart stores are led by Texas with 590 stores, followed by Florida with 386 and California with 303. These three states alone account for a significant portion of Walmart's total presence, likely reflecting their large populations and high consumer demand. Interestingly, southern states such as Texas, Florida, Georgia, and North Carolina dominate the top 10, suggesting Walmart has historically prioritized expansion in the southern United States.

```python
df_bottom_states = pd.read_csv("data/Store_Distribution_By_State_Bottom10.csv")

bottom_states = plt.barh(df_bottom_states['State'], df_bottom_states['State_Count'], color='salmon')

for bottom_state in bottom_states:
    xval = bottom_state.get_width()
    plt.text(xval, bottom_state.get_y() + bottom_state.get_height()/2, xval, ha='left', va='center')

plt.title("Bottom 10 Store Distribution by State")
plt.ylabel("State")
plt.xlabel("Number of Stores")
plt.tight_layout()
plt.show()
```

The bottom 10 states with the fewest Walmart stores are led by District of Columbia with only 2 stores, followed by Vermont with 6 and Rhode Island with 9. These states are notably smaller in size and population, which likely explains the limited Walmart presence. This constrasts sharply with Texas which has 590 stores, highlighting how population size and geography heavily influence Walmart's store distribution strategy. 

## 3. Store Type Breakdown
Question: What types of Walmart stores exist (Supercenter, Neighborhood Market, etc.), and how are they distributed?
- Visualization: Pie chart or bar chart

```python
df_store_type = pd.read_csv("data/type.csv")

plt.pie(
    df_store_type['Store_Count'],
    labels = df_store_type['Type'],
    autopct='%1.1f%%',
    colors=['steelblue', 'salmon'],
    startangle=90
)

plt.title("Walmart Store Type Breakdown")
plt.show()
```

Retail store make up 88.5% of all Walmart locations, while Wholesale accounts for the remaining 11.5%. This heavy focus on retail suggests that Walmart's primary strategy is serving everyday consumers rather than bulk buyers. However, the 601 wholesale locations still represent a significant part of Walmart's business diversity.

## 4. Top 10 States by Wholesale and Retail Store Distribution
Question: Which states have the highest concentration of Wholesale stores and how do they compare to the Retail Stores in the same state?
- Visualization: Grouped Bar Chart - one for Retail and another for Wholesale, side by side for each state.

```python
df_comparision = pd.read_csv("data/Retail_and_Wholesale_compare.csv")

x = np.arange(len(df_comparision['State']))
width = 0.35

fig, ax = plt.subplots()

retail_bars = ax.bar(x - width/2, df_comparision['Retail_Count'], width, label='Retail', color='steelblue')
wholesale_bars = ax.bar(x + width/2, df_comparision['Wholesale_Count'], width, label='Wholesale', color='salmon')

ax.set_xlabel("State")
ax.set_ylabel("Number of Stores")
ax.set_xticks(x)
ax.set_xticklabels(df_comparision['State'])
ax.legend()
plt.title("Top 10 States by Retail vs Wholesale Distribution")
plt.show()
```

Even in states with the highest Walmart presence such as Texas, Florida and California, Retail stores significantly outnumber Wholesale stores. However, it is notable that these same states also contain the highest concentration of Wholesale stores, suggesting that Walmart tends to expand its Wholesale model in states where it already has a strong Retail foundation. This could indicate a strategic approach where Walmart tests and grows its Wholesale presence in already established high performing markets.
